Introduction:

In this analysis, we aim to determine the most popular genres from year to year in a movie dataset. By examining the trends and preferences of audiences over time, we can gain insights into the evolving landscape of movie genres.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


movie_data = pd.read_csv('/content/drive/MyDrive/tmdb-movies.csv')
print(movie_data.columns)
```

```python


# Extract the relevant columns for analysis
genre_year_data = movie_data[['genres', 'release_year']]

# Group the data by release year
grouped_data = genre_year_data.groupby('release_year')

# Initialize an empty DataFrame to store the results
most_popular_genres = pd.DataFrame(columns=['Year', 'Most_Popular_Genre'])

# Iterate over each year group
for year, group in grouped_data:
    # Split the genres column into multiple genres
    genres = group['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True)

    # Count the occurrences of each genre in the group
    genre_counts = genres.value_counts()

    # Retrieve the genre(s) with the highest count
    most_popular_genres_year = genre_counts[genre_counts == genre_counts.max()].index.tolist()

    # Append the year and most popular genre(s) to the results DataFrame
    most_popular_genres = most_popular_genres.append({'Year': year, 'Most_Popular_Genre': most_popular_genres_year},
                                                     ignore_index=True)

# Print the results
print(most_popular_genres)
```

```python



# Extract the relevant columns for analysis
genre_year_data = movie_data[['genres', 'release_year']]

# Group the data by release year
grouped_data = genre_year_data.groupby('release_year')

# Initialize an empty DataFrame to store the results
most_popular_genres = pd.DataFrame(columns=['Year', 'Most_Popular_Genre'])

# Iterate over each year group
for year, group in grouped_data:
    # Split the genres column into multiple genres
    genres = group['genres'].str.split('|', expand=True).stack().reset_index(level=1, drop=True)

    # Count the occurrences of each genre in the group
    genre_counts = genres.value_counts()

    # Retrieve the genre(s) with the highest count
    most_popular_genres_year = genre_counts[genre_counts == genre_counts.max()].index.tolist()

    # Convert the list of popular genres to a string
    most_popular_genres_year_str = ', '.join(most_popular_genres_year)

    # Append the year and most popular genre(s) to the results DataFrame
    most_popular_genres = most_popular_genres.append({'Year': year, 'Most_Popular_Genre': most_popular_genres_year_str},
                                                     ignore_index=True)

# Sort the dataframe by year
most_popular_genres.sort_values('Year', inplace=True)

# Generate a bar chart to visualize the results
plt.figure(figsize=(10, 6))
plt.bar(most_popular_genres['Year'], most_popular_genres['Most_Popular_Genre'])
plt.xlabel('Year')
plt.ylabel('Most Popular Genre')
plt.title('Most Popular Genre from Year to Year')
plt.xticks(rotation=90)
plt.show()
```

Conclusion:

After analyzing the movie dataset, we found the most popular genres from year to year. The results, stored in the most_popular_genres DataFrame, provide valuable information about the changing preferences of moviegoers over time.

By examining the most_popular_genres DataFrame, we can observe the dominant genres for each year in the dataset. This insight allows us to identify the genres that captivated audiences and were most prevalent during specific periods. Understanding these trends can be useful for filmmakers, producers, and distributors in making informed decisions about movie production and marketing strategies.

Keep in mind that the conclusions drawn from this analysis are specific to the dataset used and the genres identified. It's important to consider the limitations of the dataset, such as its coverage and representativeness of the entire movie industry, as well as any biases that might be present.

```python


# Extract the relevant columns for analysis
properties = movie_data[['revenue', 'budget', 'popularity', 'vote_average', 'genres']]

# Clean the data (handle missing values, outliers, etc.)

# Calculate correlation coefficients between 'revenue' and other properties
correlation_matrix = properties.corr()

# Identify the properties with a strong positive correlation with 'revenue'
high_corr_properties = correlation_matrix['revenue'][correlation_matrix['revenue'] > 0.5].index.tolist()

# Remove 'revenue' from the list of properties
high_corr_properties.remove('revenue')

# Print the properties associated with high revenues
print("Properties associated with high revenues:")
for prop in high_corr_properties:
    print("- " + prop)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix')
plt.show()
```

Conclusion

Properties associated with movies that have high revenues:
Based on the analysis of the movie dataset, several properties are found to be associated with movies that have high revenues. These properties include:
Budget: Movies with higher budgets tend to generate higher revenues.
Popularity: Movies that are more popular among the audience tend to have higher revenues.
Vote Average: Movies with higher average ratings from viewers are also more likely to have higher revenues.
It is important to note that correlation does not imply causation. While these properties show a positive correlation with high revenues, other factors and variables may also contribute to a movie's financial success.

Visualization of the correlation matrix:
The correlation matrix provides a visual representation of the relationships between different properties and revenue. The heatmap visually represents the strength and direction of correlations. In the heatmap, warmer colors (closer to red) indicate a stronger positive correlation, while cooler colors (closer to blue) indicate a stronger negative correlation. The darker the color, the stronger the correlation.
By examining the heatmap, you can easily identify the properties that have a strong positive correlation with revenue. These properties are the ones that are closer to red in color. In this case, the properties that show a strong positive correlation with revenue include budget, popularity, and vote average.
