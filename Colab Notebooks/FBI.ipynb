````python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
````

````python
# Data Collection:
# Load the Excel file into a pandas DataFrame

data = pd.read_excel('/content/drive/MyDrive/gun-data.xlsx')
print("Columns:", data.columns)

crime_data = pd.read_csv('/content/drive/MyDrive/state_crime.csv')
print("Columns:", crime_data.columns)

census_data = pd.read_csv('/content/drive/MyDrive/census_data.csv')


````

````python
# Data Cleaning:
# Removing Duplicates
data.drop_duplicates(inplace=True)  # Remove duplicated rows if any
````

````python

# Group the data by state and calculate the sum of registrations
state_registrations = data.groupby('state')['totals'].sum()

# Sort the states based on total registration in descending order
state_registrations = state_registrations.sort_values(ascending=False)

# Plotting the data
plt.figure(figsize=(12, 6))
state_registrations.plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Total Registration')
plt.title('Gun Data Registration by State')
plt.xticks(rotation=90)
plt.show()
````

Based on the visualization on the above, it appears that Kentucky has the highest number of gun registrations compared to other states. This suggests that Kentucky has experienced significant growth in gun registrations over time, making it the state with the highest registration.
